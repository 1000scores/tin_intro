```python
from common import *
from common import plot_graph
```

```python
def plot(dataset_name, models_names, graphic_name):
    model_metrics_train = []
    model_metrics_test = []
    epochs = 1e9
    for model_name in models_names:
        cur_metrics = deserialize_metrics(dataset_name, model_name)
        model_metrics_train.append(cur_metrics['train_accuracy'])
        model_metrics_test.append(cur_metrics['test_accuracy'])
        epochs = min(epochs, cur_metrics['epoch'][-1])

    plot_graph(dataset_name, model_metrics_train, models_names, 'Train accuracy', epochs, graphic_name)
    plot_graph(dataset_name, model_metrics_test, models_names, 'Test accuracy', epochs, graphic_name)
```

** Графики для cifar10 **

```python
plot('cifar10', ['ResNet18', 'ResNet34', 'ResNet50', 'ViTResNet'], 'All models comparison')
```

```python
plot('cifar10', ['ResNet18', 'ViTResNet'], 'ViT18 vs ResNet18')
```

```python
plot('TinyImageNet', ['ViTResNet18_not_remastered', 'ViTResNet18_remastered'], 'Not remastered vs remastered')
```
